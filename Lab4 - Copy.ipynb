```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X

def idft(X):
    N = len(X)
    x = np.zeros(N, dtype=complex)
    for n in range(N):
        for k in range(N):
            x[n] += X[k] * np.exp(2j * np.pi * k * n / N)
        x[n] /= N
    return x
```

```python
N = 100
n = np.arange(N)
x = np.sin(2 * np.pi * n / N * 5) + 0.5 * np.random.randn(N)
# zeros= np.zeros(37)
# x =np.append(x,zeros)
# N = 101
# n = np.arange(N)
```

```python
X = dft(x)
X[65:101] =0
x_reconstructed = idft(X)


plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(n, x, marker='o')
plt.title("Original Signal x[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid()

# DFT Magnitude
plt.subplot(4, 1, 2)
plt.stem(np.abs(X))
plt.title("Magnitude of DFT |X[k]|")
plt.xlabel("k")
plt.ylabel("Magnitude")
plt.grid()

# DFT Phase
plt.subplot(4, 1, 3)
plt.stem(np.angle(X))
plt.title("Phase of DFT X[k])")
plt.xlabel("k")
plt.ylabel("Phase (radians)")
plt.grid()

# Reconstructed signal
plt.subplot(4, 1, 4)
plt.plot(n, x_reconstructed.real, marker='x', label='Reconstructed')
plt.title("Reconstructed Signal from IDFT ")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid()

plt.tight_layout()
plt.show()
```
