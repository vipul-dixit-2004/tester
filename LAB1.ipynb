```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
nr = np.arange(-10, 11)  # Sample indices
impulse = np.where(nr==0, 1, 0)

# Plotting the impulse signal
plt.stem(nr, impulse)
plt.title('Discrete Impulse Signal Î´[n]')
plt.xticks(np.arange(min(nr),max(nr)+1,1))
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
# Continous Unit step signal
n = np.linspace(-10, 11)  # from -10 to 10
unit_step = np.where(n >= 0, 1, 0)
# Plotting Continous Unit step signal
plt.subplot(1,2,1)
plt.plot(n, unit_step)
plt.title('Continous Unit Step Signal u[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)


# Discrete Unit Step Signa
nr = np.arange(-10, 11)
unit_step = np.where(nr >= 0, 1, 0)

# Plotting Discrete Unit Step Signa
plt.subplot(1,2,2)
plt.stem(nr, unit_step)
plt.title('Discrete Unit Step Signal u[n]')
plt.xlabel('n')
plt.xticks(np.arange(min(nr),max(nr)+1,1))
plt.ylabel('Amplitude')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
# Discrete index and continuous time
n = np.arange(-10, 11)         # Discrete time
t = np.linspace(-10, 10, 400)  # Continuous time (more points)

# Discrete ramp
discrete_ramp = np.where(n >= 0, n, 0)

# Continuous ramp
continuous_ramp = np.where(t >= 0, t, 0)

# Plotting
plt.figure(figsize=(10, 5))

# Subplot 1: Continuous ramp
plt.subplot(1, 2, 1)
plt.plot(t, continuous_ramp)
plt.title('Continuous Ramp r(t)')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)


# Subplot 1: Discrete ramp
plt.subplot(1, 2, 2)
plt.stem(n, discrete_ramp)
plt.title('Discrete Ramp r[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import thinkdsp

# Parameters
freq = 440           # Frequency of sine wave (Hz)
amp = 1.0            # Amplitude
duration = 0.01      # Duration in seconds
sample_rate = 1000   # Sampling rate (Hz)

# Step 1: Generate continuous sine wave (high resolution)
t_cont = np.linspace(0, duration, 1000)  # Very fine time steps
y_cont = amp * np.sin(2 * np.pi * freq * t_cont)

# Step 2: Create sine signal using ThinkDSP
signal = thinkdsp.SinSignal(freq=freq, amp=amp)

# Step 3: Sample the signal using ThinkDSP
wave = signal.make_wave(duration=duration, framerate=sample_rate)

# Extract sampled values
t_sampled = wave.ts
y_sampled = wave.ys

# Step 4: Plot all three: continuous, sampled, and discrete

plt.figure(figsize=(10, 5))

# Continuous sine wave
plt.plot(t_cont, y_cont, label="Continuous Sine Wave", color="blue", linewidth=1)

# Sampled wave (connected dots)
plt.plot(t_sampled, y_sampled, label="Sampled Wave", color="orange", marker='o', linestyle='--')

# Discrete samples (stem plot)
plt.stem(t_sampled, y_sampled, linefmt='green', markerfmt='go', basefmt='gray', label="Discrete Samples")

plt.title("Continuous, Sampled, and Discrete Sine Wave (440 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
