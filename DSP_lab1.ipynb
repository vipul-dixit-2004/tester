```python
import numpy as np
import matplotlib.pyplot as plt

# ================================
# Sampling settings
# ================================
Fs = 50            # Sampling frequency (Hz)
T = 1/Fs           # Sampling interval
t = np.arange(-1, 1, T)  # Time vector
n = np.arange(-20, 20, 1) # Discrete index

# ================================
# Basic Signals
# ================================
# 1. Unit Impulse (delta)
impulse_range = np.arange(-10,11)
impulse = np.where(impulse_range == 0, 1, 0)

# 2. Unit Step
step = np.where(t >= 0, 1, 0)

# 3. Ramp
ramp = np.where(t >= 0, t, 0)

# 4. Exponential (decaying)
alpha = -2
expo = np.exp(alpha * t)

# 5. Discrete Sine and Cosine
f = 2
sine_discrete = np.sin(2 * np.pi * f * n / Fs)
cosine_discrete = np.cos(2 * np.pi * f * n / Fs)

# 6. Continuous Sine and Cosine
sine_continuous = np.sin(2 * np.pi * f * t)
cosine_continuous = np.cos(2 * np.pi * f * t)


# ================================
# Plotting Basic Signals
# ================================
plt.figure(figsize=(16,14))

plt.subplot(3,3,1)
plt.stem(impulse_range, impulse)
plt.title("Unit Impulse")
plt.xlabel("n")
plt.ylabel("[n]")

plt.subplot(3,3,2)
plt.plot(t, step)
plt.title("Unit Step")
plt.xlabel("t")
plt.ylabel("u(t)")

plt.subplot(3,3,3)
plt.plot(t, ramp)
plt.title("Ramp")
plt.xlabel("t")
plt.ylabel("r(t)")

plt.subplot(3,3,4)
plt.plot(t, expo)
plt.title("Exponential")
plt.xlabel("t")
plt.ylabel("e^{t}")

plt.subplot(3, 3, 5)
plt.plot(t, sine_continuous)
plt.title('Continuous Sine')
plt.xlabel('t')
plt.ylabel('sin(t)')
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(t, cosine_continuous)
plt.title('Continuous Cosine')
plt.xlabel('t')
plt.ylabel('cos(t)')
plt.grid(True)

plt.subplot(3,2,5)
plt.stem(n, sine_discrete)
plt.title("Discrete Sine")
plt.xlabel("n")
plt.ylabel("sin()")

plt.subplot(3,2,6)
plt.stem(n, cosine_discrete)
plt.title("Discrete Cosine")
plt.xlabel("n")
plt.ylabel("cos()")

plt.tight_layout()
plt.show()

# ================================
# Continuous vs Discrete Unit Step
# ================================
plt.figure(figsize=(10, 5))

unit_step_cont = np.where(t >= 0, 1, 0)

plt.subplot(1,2,1)
plt.plot(t, unit_step_cont)
plt.title('Continuous Unit Step u(t)')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

unit_step_disc = np.where(n >= 0, 1, 0)

plt.subplot(1,2,2)
plt.stem(n, unit_step_disc)
plt.title('Discrete Unit Step u[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()

# ================================
# Continuous vs Discrete Ramp
# ================================
ramp_disc = np.where(n >= 0, n, 0)
ramp_cont = np.where(t >= 0, t, 0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(t, ramp_cont)
plt.title('Continuous Ramp r(t)')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(n, ramp_disc)
plt.title('Discrete Ramp r[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()

# ================================
# Discrete Exponentials (a^n and e^n)
# ================================
n_exp = np.arange(-10, 11)
a = float(eval(input("Enter alpha a: ")))

a_pow_n = a ** n_exp
e_pow_n = np.exp(n_exp)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.stem(n_exp, a_pow_n)
plt.title(f'Discrete Exponential: a^n (a = {a})')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(n_exp, e_pow_n)
plt.title('Discrete Exponential: e^n')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
