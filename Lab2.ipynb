# Complex function plotting
Lab2



Taking A a constant. 

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1          # amplitude
f = 5          # frequency in Hz
fs = 50        # sampling frequency in Hz
Ts = 1 / fs    # sampling period
N = 100        # number of samples

n = np.arange(N)
x = A * np.exp(1j * 2 * np.pi * f * n * Ts)  # complex exponential

plt.figure(figsize=(12, 6))

# Plot real part
plt.subplot(2, 1, 1)
plt.stem(n, x.real, basefmt=" ")
plt.grid()
plt.title('Real Part of x[n]')
plt.xlabel('n')
plt.ylabel('Re{x[n]}')

# Plot imaginary part
plt.subplot(2, 1, 2)
plt.stem(n, x.imag, basefmt=" ")
plt.title('Imaginary Part of x[n]')
plt.xlabel('n')
plt.ylabel('Im{x[n]}')
plt.grid()
plt.tight_layout()
plt.show()
````

# Taking A as a varying signal 
```A(n) = e^(αn)```
here the n will vary but we are also taking αin three different conditions

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 5          # frequency in Hz
fs = 50        # sampling frequency in Hz
Ts = 1 / fs    # sampling period
N = 100       # number of samples

alphas = [-0.05, 0, 0.05]  # varying alpha values

n = np.arange(N)

plt.figure(figsize=(12, 9))

for i, alpha in enumerate(alphas, 1):
    A = np.exp(alpha * n)             # amplitude as exponential function of n
    x = A * np.exp(1j * 2 * np.pi * f * n * Ts)
    
    plt.subplot(3, 2, 2*i - 1)
    plt.stem(n, x.real, basefmt="red")
    plt.title(f'Real Part, α = {alpha}')
    plt.grid()
    plt.xlabel('n')
    plt.ylabel('Re{x[n]}')
    
    plt.subplot(3, 2, 2*i)
    plt.stem(n, x.imag, basefmt="red")
    plt.title(f'Imaginary Part, α = {alpha}')
    plt.grid()
    plt.xlabel('n')
    plt.ylabel('Im{x[n]}')

plt.tight_layout()
plt.show()
````

# To understand the stability of the system.


````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 5
fs = 50
Ts = 1 / fs
N = 100
alphas = [-0.05, 0, 0.05]

n = np.arange(N)

plt.figure(figsize=(12, 9))

for i, alpha in enumerate(alphas, 1):
    A = np.exp(alpha * n)
    x = A * np.exp(1j * 2 * np.pi * f * n * Ts)
    
    plt.subplot(3, 1, i)
    plt.stem(n, np.abs(x), basefmt=" ")
    plt.title(f'Magnitude |x[n]| with α = {alpha}')
    plt.xlabel('n')
    plt.ylabel('|x[n]|')

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Vector length
vector_length = 10

# Impulse responses for two different vectors
impulse_vec1 = np.zeros(vector_length)
impulse_vec1[0] = 1  # Impulse at first element

impulse_vec2 = np.zeros(vector_length)
impulse_vec2[2] = 1  # Impulse at third element

# Step responses (cumulative sums of impulses)
step_vec1 = np.cumsum(impulse_vec1)
step_vec2 = np.cumsum(impulse_vec2)

# Plot impulse responses
plt.figure(figsize=(10, 4))
plt.stem(impulse_vec1, linefmt='r-', markerfmt='ro', basefmt=" ", label='Impulse Vector 1'  )
plt.stem(impulse_vec2, linefmt='b-', markerfmt='bo', basefmt=" ", label='Impulse Vector 2'  )
plt.title('Impulse Responses of Two Different Vectors')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Plot step responses
plt.figure(figsize=(10, 4))
plt.stem(step_vec1, linefmt='r-', markerfmt='ro', basefmt=" ", label='Step Vector 1'  )
plt.stem(step_vec2, linefmt='b-', markerfmt='bo', basefmt=" ", label='Step Vector 2'  )
plt.title('Step Responses of Two Different Vectors')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Vector length
vector_length = 10

# Define two different vectors (arbitrary example)
vector1 = np.array([0, 1, 2, 1, 0, 0, 0, 0, 0, 0])
vector2 = np.array([0, 0, 0, 3, 2, 1, 0, 0, 0, 0])

# Impulse responses (for demonstration, impulses inside these vectors)
# But since your vectors are arbitrary, impulse response can be thought of as the vectors themselves for now

# Step responses: cumulative sum of each vector
step_vec1 = np.cumsum(vector1)
step_vec2 = np.cumsum(vector2)

# Plot vectors
plt.figure(figsize=(10, 3))
plt.stem(vector1, linefmt='r-', markerfmt='ro', basefmt=" ", label='Vector 1'  )
plt.stem(vector2, linefmt='b-', markerfmt='bo', basefmt=" ", label='Vector 2'  )
plt.title('Signals: Two Different Vectors')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Plot impulse responses (just plot the vectors again as impulses are usually delta inputs)
plt.figure(figsize=(10, 3))
plt.stem(vector1, linefmt='r-', markerfmt='ro', basefmt=" ", label='Impulse Response Vector 1'  )
plt.stem(vector2, linefmt='b-', markerfmt='bo', basefmt=" ", label='Impulse Response Vector 2'  )
plt.title('Impulse Responses of Two Different Vectors')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Plot step responses
plt.figure(figsize=(10, 3))
plt.stem(step_vec1, linefmt='r-', markerfmt='ro', basefmt=" ", label='Step Response Vector 1'  )
plt.stem(step_vec2, linefmt='b-', markerfmt='bo', basefmt=" ", label='Step Response Vector 2'  )
plt.title('Step Responses of Two Different Vectors')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````
