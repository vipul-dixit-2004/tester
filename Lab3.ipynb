```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4,5]     
y = [-1,-2,-3,-4,-5]   


N1 = len(x)
N2 = len(y)


N = N1+N2-1


z = [0] * N


for n in range(N):
    for k in range(N1):
        if 0 <= n - k < N2:
            z[n] += x[k] * y[n - k]


p = np.convulute(x,y)
print(p)
x_indices = list(range(N1))
y_indices = list(range(N2))
z_indices = list(range(N))
print(z)


plt.figure(figsize=(12, 8))

plt.subplot(3,1,1)
plt.stem(x_indices, x)
plt.title('Input Signal x1n')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3,1,2)
plt.stem(y_indices, y)
plt.title('Input Signal x2n')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3,1,3)
plt.stem(z_indices, z)
plt.title('Convolution Result z[n] = x[n] * y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4,5]     
y = [-1,-2,-3,-4,-5]

y_flipped = y[::-1]


N1 = len(x)
N2 = len(y_flipped)


N = N1+N2-1


z = [0] * N


for n in range(N):
    for k in range(N1):
        if 0 <= n - k < N2:
            z[n] += x[k] * y_flipped[n - k]

p = np.correlate(x,y)
print(p)

x_indices = list(range(N1))
y_indices = list(range(N2))
z_indices = list(range(N))
print(z)


plt.figure(figsize=(12, 8))

plt.subplot(3,1,1)
plt.stem(x_indices, x)
plt.title('Input Signal x')
plt.xlabel('n')                                                                                                          
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3,1,2)
plt.stem(y_indices, y_flipped)
plt.title('Input Signal y')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3,1,3)
plt.stem(z_indices, z)
plt.title('Corelation Result z[n] = x[n] ** y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 0.1 * np.pi
n = np.linspace(0, 50)  # from 0 to 49

# Define the signals
x = np.sin(omega * n)
y = np.cos(omega * n)

# Compute convolution
z = np.convolve(x, y)

# Lengths for indexing
N1 = len(x)
N2 = len(y)
N = N1 + N2 - 1

# Plot signals and convolution
plt.figure(figsize=(12, 8))

plt.subplot(3,1,1)
plt.stem(n, x)
plt.title('Signal x[n] = sin(n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3,1,2)
plt.stem(n, y)
plt.title('Signal y[n] = cos(n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3,1,3)
plt.stem(np.arange(N), z)
plt.title('Convolution z[n] = x[n] * y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
